# Árbores de decisión

```python
class Node:

    def __init__(self, derecho=None, izquierdo=None, atributo=None, umbral=None, valor=None):
        self.derecho = derecho
        self.izquierdo = izquierdo
        self.atributo = atributo
        self.umbral = umbral
        self.valor = valor

    def _is_leaf(self):
        return self.valor is not None


class DecisionTree:

    def  __init__(self, max_depth=None , min_samples_split=2):
        self.max_depth = max_depth
        self.min_samples_split = min_samples_split

    def _is_finished(self):
        return True
    
    def _build_tree(self, X, y, profundidad=0):
        atributo, umbral = self._best_split(X, y)
        if atributo is None and umbral is None:
            return Node(valor=self._most_common_label(y))
        
        X_izq, y_izq, X_der, y_der = self._create_split(X, y, atributo, umbral)
        nodo_izquierdo = self._build_tree(X_izq, y_izq, profundidad + 1)
        nodo_derecho = self._build_tree(X_der, y_der, profundidad + 1)

        return Node(izquierdo=nodo_izquierdo, derecho=nodo_derecho, 
                    atributo=atributo, umbral=umbral)

    def _entropy():
        print('entropy')

    def _information_gain():
        print('information gain')

    def _create_split():
        print('create split')

    def _best_split():
        print('best split')

    def _traverse_tree():
        print('traverse tree')

    def fit(self, X, y):
        print('fit')

    def predict(self, X):
        print('predict')
    
```
